```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
```

```python
fig = plt.figure()
ax = plt.axes()
x= np.linspace(0,9,1000)
print(x)
ax.plot(x,np.sin(x))
```

```python
fig = plt.figure()
ax = plt.axes()
x= np.linspace(0,9,1000)
print(x)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
```

```python
fig = plt.figure()
ax = plt.axes()
x= np.linspace(0,9,1000)
print(x)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
ax.plot(x,np.tan(x));
```

```python
fig = plt.figure()
ax = plt.axes()
x= np.linspace(0,9,1000)
print(x)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
ax.plot(x,np.tan(x))
ax.set_title('simple plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

```python
fig = plt.figure()
ax = plt.axes()
x= np.linspace(0,9,1000)
ax.plot(x,np.sin(x), label = 'sin',linestyle = 'dashed')
ax.plot(x,np.cos(x), label = 'cos',linestyle = 'dotted')
ax.plot(x,np.tan(x), label = 'sin x+1',linestyle = 'dashdot')
ax.set_title('multiple lines');
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.legend();
plt.show()
```
